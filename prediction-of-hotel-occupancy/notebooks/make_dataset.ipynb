# Kreiranje podatkovnog okvira za praznike

```python
all_holidays = {}

for country in os.listdir(HOLIDAYS_BASE_PATH):
    if os.path.isdir(os.path.join(HOLIDAYS_BASE_PATH, country)):
        all_holidays[country] = {}
        for year_file in glob.glob(os.path.join(HOLIDAYS_BASE_PATH, country, "*.txt")):
            try:
                year = os.path.basename(year_file).split('.')[0]
                all_holidays[country][year] = []

                with open(year_file, mode='r', encoding='utf-8') as file:
                    reader = csv.reader(file)
                    headers = next(reader)
                    date_idx = headers.index("Date")
                    name_idx = headers.index("Name")
                    type_idx = headers.index("Type")

                    for row in reader:

                        date = row[date_idx]
                        date = datetime.strptime(f'{date} {year}', '%d %b %Y')
                        date= date.strftime('%Y-%m-%d') 
                        name = row[name_idx]
                        type_ = row[type_idx]
                        all_holidays[country][year].append([date, name, type_])

            except Exception as e:
                print(f"Error prcessing file: {year_file}\nError: {e}")
```

```python
holiday_portugal = pd.DataFrame(all_holidays['portugal']['2015'] + all_holidays['portugal']['2016'] + all_holidays['portugal']['2017'], columns=['date', 'name', 'type'])
holiday_portugal.to_csv(os.path.join(HOLIDAYS_BASE_PATH, 'portugal\\holidays.csv'), index=False)
```

```python
holiday_france = pd.DataFrame(all_holidays['france']['2015'] + all_holidays['france']['2016'] + all_holidays['france']['2017'], columns=['date', 'name', 'type'])
holiday_france.to_csv(os.path.join(HOLIDAYS_BASE_PATH, 'france\\holidays.csv'), index=False)
```

```python
holiday_germany = pd.DataFrame(all_holidays['germany']['2015'] + all_holidays['germany']['2016'] + all_holidays['germany']['2017'], columns=['date', 'name', 'type'])
holiday_germany.to_csv(os.path.join(HOLIDAYS_BASE_PATH, 'germany\\holidays.csv'), index=False)
```

```python
holiday_italy = pd.DataFrame(all_holidays['italy']['2015'] + all_holidays['italy']['2016'] + all_holidays['italy']['2017'], columns=['date', 'name', 'type'])
holiday_italy.to_csv(os.path.join(HOLIDAYS_BASE_PATH, 'italy\\holidays.csv'), index=False)
```

```python
holiday_spain = pd.DataFrame(all_holidays['spain']['2015'] + all_holidays['spain']['2016'] + all_holidays['spain']['2017'], columns=['date', 'name', 'type'])
holiday_spain.to_csv(os.path.join(HOLIDAYS_BASE_PATH, 'spain\\holidays.csv'), index=False)
```

```python
holiday_united_kingdom = pd.DataFrame(all_holidays['united_kingdom']['2015'] + all_holidays['united_kingdom']['2016'] + all_holidays['united_kingdom']['2017'], columns=['date', 'name', 'type'])
holiday_united_kingdom.to_csv(os.path.join(HOLIDAYS_BASE_PATH, 'united_kingdom\\holidays.csv'), index=False)
```

### Finding the intersection of the holidays

```python
holiday_intersection = holiday_portugal[['date', 'name', 'type']]
holiday_intersection = holiday_intersection.merge(holiday_france[['date']], how='inner', on='date')
holiday_intersection = holiday_intersection.merge(holiday_germany[['date']], how='inner', on='date')
holiday_intersection = holiday_intersection.merge(holiday_italy[['date']], how='inner', on='date')
holiday_intersection = holiday_intersection.merge(holiday_spain[['date']], how='inner', on='date')
holiday_intersection = holiday_intersection.merge(holiday_united_kingdom[['date']], how='inner', on='date')
```
