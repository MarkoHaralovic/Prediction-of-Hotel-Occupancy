# Kreiranje podatkovnog okvira za praznike

```python
all_holidays = {}

for country in os.listdir(HOLIDAYS_BASE_PATH):
    if os.path.isdir(os.path.join(HOLIDAYS_BASE_PATH, country)):
        all_holidays[country] = {}
        for year_file in glob.glob(os.path.join(HOLIDAYS_BASE_PATH, country, "*.txt")):
            try:
                year = os.path.basename(year_file).split('.')[0]
                all_holidays[country][year] = []

                with open(year_file, mode='r', encoding='utf-8') as file:
                    reader = csv.reader(file)
                    headers = next(reader)
                    date_idx = headers.index("Date")
                    name_idx = headers.index("Name")
                    type_idx = headers.index("Type")

                    for row in reader:

                        date = row[date_idx]
                        date = datetime.strptime(f'{date} {year}', '%d %b %Y')
                        date= date.strftime('%Y-%m-%d') 
                        name = row[name_idx]
                        type_ = row[type_idx]
                        all_holidays[country][year].append([date, name, type_])

            except Exception as e:
                print(f"Error prcessing file: {year_file}\nError: {e}")
```
